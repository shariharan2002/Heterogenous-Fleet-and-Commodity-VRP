```python
import networkx as nx
import random
```

```python
commodities=['P1','P2','P3','P4']
total_nodes=['A','B','C','D','E','F','G']
depot_nodes=['A','B']
customer_nodes=['C','D','E','F','G']

vehicle_content=0
with open(r"C:\Users\Acer Pc\Desktop\VRP OpperHeimer\OpperHeimer Code\OpperHeimer Code\Data\vessel.txt") as f:
    vehicle_content=f.read()
vehicle_content=vehicle_content.split("\n")
if vehicle_content[-1] =='' or vehicle_content[-1] =='\n':
    vehicle_content=vehicle_content[:-1]
vehicle_dict={}
for item in vehicle_content:
    item=item.split(":")
    vehicle_cap=int(item[1].split(",")[0].split("(")[1].strip())
    vehicle_cost=int(item[1].split(",")[1].split(")")[0].strip())
    vehicle_dict[item[0].strip()]=[vehicle_cap,vehicle_cost]

capacity_content=0
with open(r"C:\Users\Acer Pc\Desktop\VRP OpperHeimer\OpperHeimer Code\OpperHeimer Code\Data\capacity.txt") as f:
    capacity_content=f.read()
capacity_content=capacity_content.split("\n")

if capacity_content[-1]==" " or capacity_content[-1]=="\n":
    capacity_content=capacity_content[:-1]

capacity_dict={}
for item in capacity_content:
    item=item.split(":")
    depot_id=item[0].split(",")[0].split("(")[1].strip()
    product_id=item[0].split(",")[1].split(")")[0].strip()
    depot_cap=int(item[1].strip())
    capacity_dict[(depot_id,product_id)]=depot_cap

time_content=0
with open(r"C:\Users\Acer Pc\Desktop\VRP OpperHeimer\OpperHeimer Code\OpperHeimer Code\Data\time.txt") as f:
    time_content=f.read()
time_content=time_content.split("\n")

if time_content[-1]==" " or time_content[-1]=="\n":
    time_content=time_content[:-1]
mapper={0:'A',1:'B',2:'C',3:'D',4:'E',5:'F',6:'G'}
time_dict={}
for item in time_content:
    item=item.split(":")
    source_depot=int(item[0].split(",")[0].split("(")[1].strip())
    destination_depot=int(item[0].split(",")[1].split(")")[0].strip())
    time_to_travel=int(item[1].strip())
    time_dict[(mapper[source_depot],mapper[destination_depot])]=time_to_travel
    time_dict[(mapper[destination_depot],mapper[source_depot])]=time_to_travel

demand_content=0
with open(r"C:\Users\Acer Pc\Desktop\VRP OpperHeimer\OpperHeimer Code\OpperHeimer Code\Data\demand.txt") as f:
    demand_content=f.read()
demand_content=demand_content.split("\n")

if demand_content[-1]==" " or demand_content[-1]=="\n":
    demand_content=demand_content[:-1]

demand_dict={}
for item in demand_content:
    item=item.split(":")
    depot_id=item[0].split(",")[0].split("(")[1].strip()
    product_id=item[0].split(",")[1].split(")")[0].strip()
    demand_for_product=int(item[1].strip())
    demand_dict[(depot_id,product_id)]=demand_for_product
for node in depot_nodes:
    for commodity in commodities:
        demand_dict[(node,commodity)]=0
```

```python
normal_arcs_dict={}


for k,v in vehicle_dict.items():
    for c in range(len(commodities)):
        normal_arcs_dict[(k,commodities[c])]=[]   
        for i in range(len(total_nodes)):
            for j in range(len(total_nodes)):
                if total_nodes[i]!=total_nodes[j]   and (total_nodes[i] not in depot_nodes  or total_nodes[j] not in depot_nodes):
                    if demand_dict[(total_nodes[i],commodities[c])] + demand_dict[(total_nodes[j],commodities[c])] <= v[0]:
                        normal_arcs_dict[(k,commodities[c])].append((total_nodes[i],total_nodes[j]))


```

```python
replenishment_arcs_dict={}
for keys,values in vehicle_dict.items():
    for c in range(len(commodities)):
        replenishment_arcs_dict[(keys,commodities[c])]=[]   
        for i in range(len(total_nodes)):
            for j in range(len(total_nodes)):
                for k in range(len(total_nodes)):
                    if (total_nodes[i] in depot_nodes) and (total_nodes[j] in customer_nodes) and (total_nodes[k] in customer_nodes) and (total_nodes[j]!=total_nodes[k]) and (demand_dict[(total_nodes[j],commodities[c])]<= values[0]) and (demand_dict[(total_nodes[k],commodities[c])]<= values[0]):
                        replenishment_arcs_dict[(keys,commodities[c])].append((total_nodes[i],total_nodes[j],total_nodes[k]))
```

```python
normal_arcs_cost_dict={}
for keys,values in vehicle_dict.items():
    for c in range(len(commodities)):
        normal_arcs_cost_dict[(keys,commodities[c])]={}
        for edges in normal_arcs_dict[(keys,commodities[c])]:
            normal_arcs_cost_dict[(keys,commodities[c])][edges]=time_dict[edges]*values[1]
```

```python
replenishment_arcs_cost_dict={}
for keys,values in vehicle_dict.items():
    for c in range(len(commodities)):
        replenishment_arcs_cost_dict[(keys,commodities[c])]={}
        for edges in replenishment_arcs_dict[(keys,commodities[c])]:
            replenishment_arcs_cost_dict[(keys,commodities[c])][edges]=normal_arcs_cost_dict[(keys,commodities[c])][(edges[1],edges[0])]+normal_arcs_cost_dict[(keys,commodities[c])][(edges[0],edges[2])]
```

```python
G=nx.Graph()
for node in total_nodes:
    G.add_node(node)
for edge in time_dict.keys():
    if edge[0]!=edge[1]:
        G.add_edge(edge[0],edge[1],weight=time_dict[edge])

```

```python
from itertools import combinations,permutations

global_paths_set=set()
all_paths_list=[]
roundtrips_dict={}

for depot in depot_nodes:
    roundtrips_dict[depot]=set()
    for node in total_nodes:
        for path in list(nx.all_simple_paths(G,depot,node)):
            if depot not in path[1:]:
                temp=path.copy()
                temp.append(depot)
                temp=tuple(temp)
                roundtrips_dict[depot].add((temp))

for permutation in permutations(depot_nodes, 2):
    source_depot=permutation[0]
    destination_depot=permutation[1]
    for path in list(nx.all_simple_paths(G,source=source_depot,target=destination_depot)):
        temp=tuple(path)
        all_paths_list.append(path)
        global_paths_set.add(temp)

for depot in depot_nodes:
    for path in roundtrips_dict[depot]:
        all_paths_list.append(depot)
        global_paths_set.add(path)



        
        
```

```python
feasible_vehicle_dict={}
for cust_node in customer_nodes:
    for c in range(len(commodities)):
        feasible_vehicle_dict[(cust_node,commodities[c])]=[]
        for vehicle_index,v1 in vehicle_dict.items():
            if demand_dict[(cust_node,commodities[c])]<=v1[0]:
                feasible_vehicle_dict[(cust_node,commodities[c])].append(vehicle_index)
    

```

```python


z_decision_variable_indices=[]
objective_function_values={}
for vehicle_index,v1 in vehicle_dict.items():
    for c in range(len(commodities)): 
        for path_index,v2 in feasible_path_dict[(vehicle_index,commodities[c])].items():
            z_decision_variable_indices.append((vehicle_index,path_index,commodities[c]))
            objective_function_values[(vehicle_index,path_index,commodities[c])]=feasible_path_dict[(vehicle_index,commodities[c])][path_index]['Cost']

from docplex.mp.model import Model

vrp_model = Model('Heterogenous Fleet MDMT-VRP')
z=vrp_model.binary_var_dict(z_decision_variable_indices,name='z')

# a_decision_variable_indices=[]

# for vehicle_index,v1 in vehicle_dict.items():
#     for depot_node in depot_nodes:
#         a_decision_variable_indices.append((vehicle_index,depot_node))

# a=vrp_model.continuous_var_dict(a_decision_variable_indices,name='a')


obj_fn=0
for vehicle_index,v1 in vehicle_dict.items():
    for c in range(len(commodities)):
        for path_index,v2 in feasible_path_dict[(vehicle_index,commodities[c])].items():
            obj_fn+=objective_function_values[(vehicle_index,path_index,commodities[c])]*z[(vehicle_index,path_index,commodities[c])]



#obj_fn=sum(objective_function_values[(v_index,p_index)]*x[(v_index,p_index)] for p_index,v2 in feasible_path_dict[v_index].items() for v_index,v1 in vehicle_dict.items() ))

# obj_fn = sum(
#     objective_function_values[(v_index, p_index)] * x[(v_index, p_index)]
#     for v_index, v1 in vehicle_dict.items()
#     for p_index, v2 in feasible_path_dict[v_index].items()
    
# )

vrp_model.set_objective("min",obj_fn)

for j in customer_nodes:
    for c in range(len(commodities)):
        c_once=0
        for vehicle_index,v1 in vehicle_dict.items():
            if vehicle_index in feasible_vehicle_dict[(j,commodities[c])]:         
                for path_index,v2 in feasible_path_dict[(vehicle_index,commodities[c])].items():
                    if j in feasible_path_dict[(vehicle_index,commodities[c])][path_index]['CustomerSet']:
                        c_once+=z[(vehicle_index,path_index,commodities[c])]
            else:
                continue
        if c_once:
            #print("BRO")
            vrp_model.add_constraint(c_once==1,'cvo') #customer visit once(cvo)



for i in depot_nodes:
    for c in range(len(commodities)):
        depot_cap_constraint=0
        for vehicle_index,v1 in vehicle_dict.items():
            for path_index,v2 in feasible_path_dict[(vehicle_index,commodities[c])].items():
                for node_index in range(len(feasible_path_dict[(vehicle_index,commodities[c])][path_index]['Path'])-1):
                    if feasible_path_dict[(vehicle_index,commodities[c])][path_index]['Path'][node_index] in depot_nodes and feasible_path_dict[(vehicle_index,commodities[c])][path_index]['Path'][node_index] == i:
                        depot_cap_constraint+=(feasible_path_dict[(vehicle_index,commodities[c])][path_index]['ConsumptionTracker'][vehicle_index,path_index,commodities[c],node_index]*z[(vehicle_index,path_index,commodities[c])]) 
        if depot_cap_constraint:
            #print("HELLO",i,commodities[c])
            vrp_model.add_constraint(depot_cap_constraint <= capacity_dict[(i,commodities[c])], 'dcc') #depot capacity constraint(dcc)

#------- --------------------------------!!!!!!!!!!!!!!!!!!!!!!!!!!

for i in depot_nodes:
    for vehicle_index, v1 in vehicle_dict.items():
        for c in range(len(commodities)):
            lhs=0
            for j in depot_nodes:
                for path_index, v2 in feasible_path_dict[(vehicle_index,commodities[c])].items():
                    if feasible_path_dict[(vehicle_index,commodities[c])][path_index]['Start']==i and feasible_path_dict[(vehicle_index,commodities[c])][path_index]['End']==j:
                        lhs+=z[(vehicle_index,path_index,commodities[c])]
            
            rhs=0
            for j in depot_nodes:
                for path_index, v2 in feasible_path_dict[(vehicle_index,commodities[c])].items():
                    if feasible_path_dict[(vehicle_index,commodities[c])][path_index]['Start']==j and feasible_path_dict[(vehicle_index,commodities[c])][path_index]['End']==i:
                        rhs+=z[(vehicle_index,path_index,commodities[c])]

            #print("hhh",i,vehicle_index,commodities)
            vrp_model.add_constraint(lhs==rhs, 'dbc') #depot balance constraint(dbc)
            
# #NEW SET OF CONSTRAINTS !!!!

# for vehicle_index,v1 in vehicle_dict.items():
#     for path_index,v2 in feasible_path_dict[vehicle_index].items():
#         lhs_dcs=0

#         for 

sol=vrp_model.solve()


    

sol_values=sol.get_value_dict(z)
for k,v in sol_values.items():
    if v==1:
        print(True,k)


sol.get_objective_value()


```
